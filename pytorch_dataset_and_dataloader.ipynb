```python
class CustomDataset(Dataset):
  def __init__(self):
    data = np.loadtxt('wine.csv',
                      delimiter = ',',
                      dtype = np.float32,
                      skiprows = 1)

    self.x_data = torch.from_numpy(data[:, 1:])
    self.y_data = torch.from_numpy(data[:, [0]])
    self.samples = data.shape[0]

  def __len__(self):
    return self.samples

  def __getitem__(self, index):
    return self.x_data[index], self.y_data[index]
```

```python
dataset = CustomDataset()
dataloader = torch.utils.data.DataLoader(dataset = dataset,
                                         batch_size = 4,
                                         shuffle = True,
                                         num_workers = 2)
```

```python
# Verifying that CustomDataset Class is working correctly

first_data = dataset[0]
features, labels = first_data
print(features, labels)
```

```python
# Verifying that Dataloader is working correctly

for features, labels in dataloader:
  print(features, labels)
  break
```

```python
# Dummy Training Loop

epochs = 2
total_samples = len(dataset)
n_iterations = math.ceil(total_samples / 4)

# print(total_samples, n_iterations)
for epoch in range(epochs):
  for i, (input, output) in enumerate(dataloader):
    # Training steps
    if (i + 1) % 5 == 0:
      print(f'epoch : {epoch + 1}/{epochs}, steps : {i+1}/{n_iterations}, inputs : {input.shape}')
```
