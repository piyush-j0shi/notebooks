```python
import torch
import numpy as np
```

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)

print(f"data : {x_data}")
```

```python
np_arry = np.array(data)
x_np = torch.from_numpy(np_arry)

print(f"from_numpy : {x_np}")
```

```python
x_ones = torch.ones_like(x_data)
print(f"ones tensors : {x_ones}")

x_rand = torch.rand_like(x_data, dtype = torch.float32)
print(f"random : {x_rand}")
```

```python
shape = (2, 3,)

rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"random : {rand_tensor}\n")
print(f"ones : {ones_tensor}\n")
print(f"zeors : {zeros_tensor}\n")
```

```python
tensor = torch.rand(3, 4,)
# print(tensor)

print(f"shape of tensor : {tensor.shape}")
print(f"datatype of tensor : {tensor.dtype}")
print(f"device tensor is stored on : {tensor.device}")
```

```python
if torch.accelerator.is_available():
  tensor = torch.to(torch.accelerator.current_accelerator())
```

```python
tensor = torch.ones(4, 4)

print(f"first row : {tensor[0]}")
print(f"first column : {tensor[:, 0]}")
print(f"last column : {tensor[..., -1]}")

tensor[:, 1] = 0
print(tensor)
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim = 1)
print(t1)

t2 = torch.cat([tensor, tensor, tensor], dim = 0)
print(t2)

t3 = t1 = torch.cat([tensor, tensor, tensor], dim = 1)
print(t3.add(5))
```

```python
t = torch.ones(5)
print(f"t : {t}")

n = t.numpy()
print(f"n : {n}")
```

```python
t.add_(1)

print(f"t : {t}")
print(f"n : {n}")
```

```python
n = np.ones(5)
t = torch.from_numpy(n)

print(f"t : {t}")
print(f"n : {n}")
```

```python
n = np.ones(5)
t = torch.from_numpy(n)

np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
