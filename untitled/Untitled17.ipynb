```python
import pandas as pd
import numpy as np
import tensorflow as tf

from datasets import load_dataset
from transformers import BertTokenizer
from transformers import TFBertForSequenceClassification
```

```python
data = load_dataset("mteb/banking77")
```

```python
train_data = data['train']
test_data = data['test']
```

```python
print(data)
```

```python
print(train_data)
print(test_data)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def example_tokenize(example):
  return tokenizer(example['text'], truncation = True, padding = 'max_length', max_length = 128)
```

```python
encoded_train = train_data.map(example_tokenize, batched = True)
encoded_test = test_data.map(example_tokenize, batched = True)
```

```python
def convert_to_tf_dataset(encoded_dataset):
    features = {x: encoded_dataset[x] for x in tokenizer.model_input_names}
    tf_dataset = tf.data.Dataset.from_tensor_slices((features, encoded_dataset['label']))
    return tf_dataset
```

```python
tf_train_dataset = convert_to_tf_dataset(encoded_train)
tf_test_dataset = convert_to_tf_dataset(encoded_test)
```

```python
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels = 75)
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-5),
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    metrics = ['accuracy']
)
```
