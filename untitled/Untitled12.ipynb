```python
import nltk
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
from nltk.stem import PorterStemmer
from nltk.corpus import twitter_samples
```

```python
# Lets select one tweet on which we want to perform tokenie, steamming, lowercasing,remove stopwords and panctuations

tweet = all_positive_tweets[2277]
```

```python
tweet
```

```python
# Remove hyperlinks, handles, twitter marks and styles

tweet_0 = re.sub(r'^RT[\s]+', '', tweet)
tweet_0 = re.sub(r'https?://[^\s\n\r]+', '', tweet_0)
tweet_0 = re.sub(r'#', '', tweet_0)

print(tweet_0)
```

```python
tokenizer = TweetTokenizer(preserve_case = False, reduce_len = True, strip_handles = True)

tweet_tokens = tokenizer.tokenize(tweet_0)
print(tweet_tokens)
```

```python
nltk.download('stopwords')
stopwords_english = stopwords.words('english')

print(stopwords_english)
print(string.punctuation)
```

```python
clean_tweets = []

for word in tweet_tokens:
  if (word not in stopwords_english and
    word not in string.punctuation):
    clean_tweets.append(word)

# print(clean_tweets)
clean_tweets
```

```python
steam_tweets = []

streammer = PorterStemmer()

for word in clean_tweets:
  steam_word = streammer.stem(word)
  steam_tweets.append(steam_word)

steam_tweets
```

```python
def processed_tweets(tweet_1):
  tweet_2 = re.sub(r'^RT[\s]+', '', tweet_1)
  tweet_2 = re.sub(r'https?://[^\s\n\r]+', '', tweet_2)
  tweet_2 = re.sub(r'#', '', tweet_2)

  tokenizer_0 = TweetTokenizer(preserve_case = False, reduce_len = True, strip_handles = True)
  tokenize_tweets = tokenizer_0.tokenize(tweet_2)

  # nltk.download('stopwords')

  stopwords_english_0 = stopwords.words('english')
  clean_tweet_0 = []

  for word in tokenize_tweets:
    if (word not in stopwords_english_0 and
        word not in string.punctuation):
      clean_tweet_0.append(word)

  stream = PorterStemmer()
  stem_tweets_0 = []

  for word in clean_tweet_0:
    steam_word_0 = stream.stem(word)
    stem_tweets_0.append(steam_word_0)

  return stem_tweets_0
```

```python
tweets = all_positive_tweets[2277]
tweets
```

```python
processed_tweets(tweets)
```

```python
all_proccessed_positive_tweets = []

for tweet in all_positive_tweets:
  processed_tweet = processed_tweets(tweet)
  all_proccessed_positive_tweets.append(processed_tweet)
```
