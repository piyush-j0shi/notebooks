```python
from pathlib import Path
from llama_index.readers.file import UnstructuredReader
import nltk

# Download NLTK data if not already downloaded
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# List of years
years_0 = [2022, 2021, 2020, 2019]

# Create an UnstructuredReader object
loader_0 = UnstructuredReader()

# Initialize dictionaries and lists
doc_set_0 = {}
all_docs_0 = []

# Loop through each year
for year in years_0:
    try:
        # Load data from HTML files
        year_docs = loader_0.load_data(
            file=Path(f"./data/UBER/UBER_{year}.html"), split_documents=False
        )

        # Insert year metadata into each document
        for d in year_docs:
            if isinstance(d, dict):
                d['metadata'] = {"year": year}
            else:
                if hasattr(d, 'metadata'):
                    d.metadata = {"year": year}
                else:
                    # If metadata cannot be set directly, handle accordingly
                    d = {'doc': d, 'metadata': {"year": year}}

        # Add to doc_set and all_docs
        doc_set_0[year] = year_docs
        all_docs_0.extend(year_docs)

    except Exception as e:
        print(f"Error loading or processing file for year {year}: {e}")

# Output information about the documents loaded
print(f"Loaded documents for years: {list(doc_set_0.keys())}")
print(f"Total documents loaded: {len(all_docs_0)}")
```

```python
from transformers import pipeline

# Initialize the chatbot pipeline
chatbot = pipeline("text-generation", model="gpt2")

# Define a function to get responses from the chatbot with custom data
def chat_with_bot(user_input):
    # Combine custom data and user input
    context = ' '.join([d.text for d in all_docs_0])  # Adjust this if text is accessed differently

    # Truncate context if necessary
    max_input_length = 400  # Adjust this based on model limitations
    if len(context) > max_input_length:
        context = context[-max_input_length:]  # Keep only the last part of the context

    # Refine the prompt to be more specific
    prompt = f"Context: {context}\n\nUser: {user_input}\n\nChatbot:"

    # Generate response
    response = chatbot(
        prompt,
        max_length=500,  # Adjust max_length as needed
        truncation=True,  # Ensure truncation is enabled
        return_full_text=True
    )
    return response[0]['generated_text']

# Example interaction
user_input = "What were some of the legal proceedings against Uber in 2022?"
response = chat_with_bot(user_input)
print("Chatbot:", response)
```

```python
print("Chatbot is ready to chat. Type 'exit', 'quit', or 'bye' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit', 'bye']:
        print("Chatbot: Goodbye!")
        break
    response = chat_with_bot(user_input)
    print("Chatbot:", response)
```
