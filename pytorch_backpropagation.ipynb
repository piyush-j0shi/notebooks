```python
import torch
import numpy as np
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad = True)
```

```python
print(x, y, w)
```

```python
# Forward pass
y_hat = w * x
s = y_hat - y
loss = s ** 2
```

```python
print(loss)
```

```python
# Backward pass
loss.backward()
print(w.grad)
```

```python
value1 = torch.tensor(1.0)
value2 = torch.tensor(2.0)
weights = torch.tensor(1.0, requires_grad = True)
```

```python
print(value1, value2, weights)
```

```python
yhat = weights * value1
intermediate_value = yhat - value2
loss = intermediate_value ** 2
```

```python
print(loss)
```

```python
loss.backward()
print(weights.grad)
```
