```python
class Dataset_0(Dataset):
  def __init__(self, csv_file, data_dir, transforms = True):
    self.data_dir = data_dir
    self.transforms = transforms
    csv_path = os.path.join(self.data_dir, csv_file)
    self.data_name = pd.read_csv(csv_path)
    self.len = self.data_name.shape[0]

  def __len__(self):
    return self.len

  def __getitem__(self, idx):
    image_name = os.path.join(self.data_dir, self.data_name.oloc[idx, 1])
    image = Image.open(image_name)
    y = self.data_name[idx, 0]
    if self.transform:
      image = self.transform(image)

    return image, y
```

```python
new_transform = transforms.Compose([transforms.CenterCrop(20), transforms.ToTensor()])
dataset = Dataset_0(csv_file, data_dir = directory, transforms = new_transform)
```
