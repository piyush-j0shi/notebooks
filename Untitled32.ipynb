```python
from google.colab import auth
auth.authenticate_user()

from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseDownload
from google.colab import files
import os, io, shutil
from googleapiclient.errors import HttpError

# Setup
drive_service = build('drive', 'v3')
download_dir = '/content/colab_notebooks'
os.makedirs(download_dir, exist_ok=True)

# Search for all .ipynb notebooks in Google Drive
query = "name contains '.ipynb'"
results = drive_service.files().list(q=query, pageSize=1000, fields="files(id, name, mimeType)").execute()
items = results.get('files', [])

if not items:
    print("No notebooks found.")
else:
    print(f"Found {len(items)} notebooks. Attempting to download valid ones...")

    downloaded = 0
    for item in items:
        file_id = item['id']
        file_name = item['name']

        try:
            request = drive_service.files().get_media(fileId=file_id)
            filepath = os.path.join(download_dir, file_name)

            with io.FileIO(filepath, 'wb') as fh:
                downloader = MediaIoBaseDownload(fh, request)
                done = False
                while not done:
                    status, done = downloader.next_chunk()

            print(f"Downloaded: {file_name}")
            downloaded += 1

        except HttpError as error:
            if error.resp.status == 403:
                print(f"Skipping '{file_name}' (403 Forbidden - Not downloadable)")
            else:
                print(f"Error downloading '{file_name}': {error}")

    print(f"\nFinished. Successfully downloaded {downloaded}/{len(items)} notebooks.")

# Zip the downloaded folder
zip_path = '/content/colab_notebooks_backup.zip'
shutil.make_archive('/content/colab_notebooks_backup', 'zip', download_dir)

# Prompt local download
files.download(zip_path)
```
