```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
data['Gender'] = data['Gender'].map(lambda x:1 if x=='Male' else 0)

le = LabelEncoder()
data['Geography'] = le.fit_transform(data['Geography'])

train, test = train_test_split(data, test_size = 0.2, random_state=42)
```

```python
train.shape
train.info()
train.head()
```

```python
my_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])
```

```python
train_features = train.drop(['Exited'], axis=1)
train_labels = train['Exited'].copy()

test_features = test.drop(['Exited'], axis = 1)
test_labels = test['Exited'].copy()
```

```python
train_features.shape
# train_labels.shape
```

```python
train_features = my_pipeline.fit_transform(train_features)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([

    # Input shape or input_dim = total no of columns in training data OR train_features.shape[1]
    # input_dim = 10(For this case bcz in this dataset we have 10 columns in our training data OR it can be input_dim = train_features.shape[1])
    # beacuse train_features.shape[1] is equal to "no of columns in training data".
    # And for some additional information train_features.shape[1] is equal to **"no of rows in tarining data"**
    # print(train_features.shape[1])
    # print(train_features.shape[0])
    # During training a numerical type data you can not write input_shape you always have to write input_dim
    # And if you want to use input_shape then you have to use this in the end like this input_shape = (train_features.shape[1],)

    tf.keras.layers.Dense(64, input_dim = train_features.shape[1], activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

model.summary()
```

```python
model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
history = model.fit(
    train_features,
    train_labels,
    epochs = 100,
    batch_size = 40,
    validation_data = (test_features, test_labels)
)
```
