```python
import os
import shutil
from sklearn.model_selection import train_test_split

def split_dataset(data_dir, t_dir, v_dir, val_size = 0.2):
  if not os.path.exists(t_dir):
    os.makedirs(t_dir)
  if not os.path.exists(v_dir):
    os.makedirs(v_dir)

  folders = os.listdir(data_dir)
  for folder in folders:
    folder_path = os.path.join(data_dir, folder)
    if not os.path.isdir(folder_path):
      continue

    images = os.listdir(folder_path)
    t_images, v_images = train_test_split(images, test_size = val_size, random_state = 42)

    t_folder = os.path.join(t_dir, folder)
    v_folder = os.path.join(v_dir, folder)

    if not os.path.exists(t_folder):
      os.makedirs(t_folder)
    if not os.path.exists(v_folder):
      os.makedirs(v_folder)

    for image in t_images:
      shutil.copy(os.path.join(folder_path, image), os.path.join(t_folder, image))
    for image in v_images:
      shutil.copy(os.path.join(folder_path, image), os.path.join(v_folder, image))

data_dir = 'shapes'
t_dir = 'train'
v_dir = 'valid'

split_dataset(data_dir, t_dir, v_dir, val_size = 0.2)
```
