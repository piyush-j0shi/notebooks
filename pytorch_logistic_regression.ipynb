```python
import torch
import numpy as np
import torch.nn as nn
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = datasets.load_breast_cancer()
x, y = data.data, data.target
```

```python
n_samples, n_features = x.shape
print(n_samples, n_features)

input_size = n_features
output_size = 1 # In Classification tesks output size is always No Of Classes or no of labels
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
standard_scaler = StandardScaler()

x_train = standard_scaler.fit_transform(x_train)
x_test = standard_scaler.transform(x_test)
```

```python
x_train = torch.from_numpy(x_train.astype(np.float32))
x_test = torch.from_numpy(x_test.astype(np.float32))

y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
```

```python
y_train = y_train.view(y_train.shape[0], 1)
y_test = y_test.view(y_test.shape[0], 1)
```

```python
class LogisticRegression(nn.Module):
  def __init__(self, input_dimension, output_dimension):
    super(LogisticRegression, self).__init__()
    self.linear = nn.Linear(input_dimension, output_dimension)

  def forward(self, x):
    y_predicted = torch.sigmoid(self.linear(x))
    return y_predicted

model = LogisticRegression(input_size, output_size)
```

```python
learning_rate = 1e-2
loss_function = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
epochs = 100
for epoch in range(epochs):
  y_predicted = model(x_train)
  loss = loss_function(y_predicted, y_train)

  loss.backward() # Loss will not chnage if we do not use this line.
  optimizer.step() # Loss will not change if we do not use this line.
  optimizer.zero_grad()

  if (epoch + 1) % 10 == 0:
    print(f'epoch : {epoch + 1}, loss : {loss.item():.4f}')

with torch.no_grad():
  y_predicted = model(x_test)
  # print(y_predicted)

  y_predicted_correct = y_predicted.round()
  # print(y_predicted_correct)

  accuracy = y_predicted_correct.eq(y_test).sum() / len(y_test)

  # print(y_predicted_correct.eq(y_test))
  # print( y_predicted_correct.eq(y_test).sum())
  # print(len(y_test))

  print(f'accuracy : {accuracy.item():.4f}')
```
