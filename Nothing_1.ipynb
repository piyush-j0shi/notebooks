***Dataset Class And Transforms***

```python
import torch
from torch.utils.data import Dataset
from torchvision import transforms
```

```python
class toy_set(Dataset):
  def __init__(self, length = 100, transform = None):
    self.x = 2*torch.ones(length, 2)
    self.y = torch.ones(length, 1)
    self.len = length
    self.transform = transform

  def __getitem__(self, index):
    sample = self.x[index], self.y[index]
    if self.transform:
      sample = self.transform(sample)
    return sample

  def __len__(self):
    return self.len
```

```python
dataset = toy_set()

print(dataset[0])
print(len(dataset))
```

```python
for i in range(3):
  print(f"X for {i} = {dataset[i][0]}")
  print(f"Y for {i} = {dataset[i][1]}")
  print()
```

```python
class add_mult(object):
  def __init__(self, addx = 1, muly = 1):
    self.addx = addx
    self.muly = muly

  def __call__(self, sample):
    x = sample[0]
    y = sample[1]
    x = x + self.addx
    y = y * self.muly
    sample = x, y
    return sample

a_m = add_mult()
```

```python
x_, y_ = a_m(dataset[0])

print(f"X Before Transforms = {dataset[0][0]}")
print(f"Y Before Transforms = {dataset[0][1]}")
print()
print(f"X After Transforms = {x_}")
print(f"Y After Transforms = {y_}")
```

```python
# Series Of Transforms
class mult(object):
  def __init__(self, mult = 100):
    self.mult = mult

  def __call__(self, sample):
    x = sample[0]
    y = sample[1]
    x = x * self.mult
    y = y * self.mult
    sample = x, y
    return sample

m = mult()
```

```python
# from torchvision import transforms
data_transforms = transforms.Compose([add_mult(), mult()])
x_, y_ = data_transforms(dataset[0])

dataset__ = toy_set(length = 50, transform = data_transforms)
```
