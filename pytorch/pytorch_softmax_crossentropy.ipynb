```python
import torch
import numpy as np
import torch.nn as nn
```

```python
# Custom Function
def softmax(x):
  probability = np.exp(x) / np.sum(np.exp(x), axis = 0)
  return probability

x = np.array([2.0, 1.0, 0.1])
outputs = softmax(x)
print(f'outputs : {outputs}')
```

```python
# Using Pytorch's Softmax Function
x = torch.tensor([2.0, 1.0, 0.1])
outputs = torch.softmax(x, dim = 0)
print(f'outputs : {outputs}')
```

```python
def cross_entropy(actual, predicted):
  loss = -np.sum(actual * np.log(predicted))
  return loss

actual = np.array([1, 0, 0])
good_predicted = np.array([0.7, 0.2, 0.1])
bad_predicted = np.array([0.1, 0.3, 0.6])

loss_1 = cross_entropy(actual, good_predicted)
loss_2 = cross_entropy(actual, bad_predicted)

print(f'loss_1 Numpy : {loss_1 : .4f}')
print(f'loss_2 Numpy : {loss_2 : .4f}')
```

```python
actual = torch.tensor([0]) # One Class
actual_1 = torch.tensor([2, 0, 1]) # All Classes

good_predcited = torch.tensor([[2.0, 1.0, 0.1]])
bad_predicted = torch.tensor([[0.5, 2.0, 0.3]])

good_predcited_1 = torch.tensor([[2.0, 1.0, 0.1], [0.3, 1.0, 0.1], [0.1, 0.2, 1.0]])
bad_predicted_1 = torch.tensor([[0.5, 2.0, 0.3], [1.5, 0.2, 0.3], [0.5, 0.2, 1.3]])

loss = nn.CrossEntropyLoss()
loss_1 = loss(good_predcited, actual)
loss_2 = loss(bad_predicted, actual)

loss_3 = loss(good_predcited_1, actual_1)
loss_4 = loss(bad_predicted_1, actual_1)

print(f'loss_1 Pytorch : {loss_1.item()}')
print(f'loss_2 Pytorch : {loss_2.item()}')

print(f'loss_3 Pytorch : {loss_3.item()}')
print(f'loss_4 Pytorch : {loss_4.item()}')

# i1, prediction_1 = torch.max(good_predcited, 1)
# i2, prediction_2 = torch.max(bad_predicted, 1)

'''
i1 will give you the indices of the maximum values (class indices) along the specified dimension
prediction_1 will give you the maximum values themselves (the actual prediction scores or probabilities) for each example in the batch.
That's why we use _ at the place of i1 and i2
'''

_, prediction_1 = torch.max(good_predcited, 1)
_, prediction_2 = torch.max(bad_predicted, 1)

print(f'prediction_1 : {prediction_1}')
print(f'prediction_2 : {prediction_2}')

# print(i1)
# print(i2)
```
