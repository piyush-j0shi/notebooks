```python
import numpy as np
import torch
```

```python
# Y = W * X
# Y = 2 * X

"""
# Step : 1
X = np.array([1, 2, 3, 4], dtype = np.float32)
Y = np.array([2, 4, 6, 8], dtype = np.float32)

# Initial weights W = 0.0
W = 0.0
"""

# Step : 2
X = torch.tensor([1, 2, 3, 4], dtype = torch.float32)
Y = torch.tensor([2, 4, 6, 8], dtype = torch.float32)

# Initial weights W = 0.0
W = torch.tensor(0.0, dtype = torch.float32, requires_grad = True)
```

```python
# Forward pass means Linear Regression Model
def forward_pass(w, x):
  return w * x
```

```python
# Loss Function
def loss_function(y, y_predicted):
  return ((y_predicted - y) ** 2).mean()
```

```python
"""
#  Step : 1
# Gradient and Backpropagation
def gradient(x, y, y_predicted):
  return np.dot(2 * x, (y_predicted - y)).mean()
"""

# In step : 2 we use .backward() and .grad() functions instead of doing it manually
print("working")
```

```python
# Here 0 is weights value and 5 is input for example
print(f'prediction before training on digit 5: f(5) = {forward_pass(W, 5) : .3f}')
```

```python
# Training
learning_rate = 0.01
epochs = 10

for epoch in range(epochs):
  # Predictions = forward_pass
  Y_PRED = forward_pass(W, X)
  # print(Y_PRED)

  # loss
  loss = loss_function(Y, Y_PRED)

  # Gradient Calculation

  """
  # Step : 1
  def gradient_calculation(x, y, y_predicted):
	   return np.dot(2 * x, (y_predicted - y)).mean()
  # print(grad)
  """

  # Step : 2
  dw = loss.backward()

  with torch.no_grad():
    W -= learning_rate * W.grad

  W.grad.zero_()

  if epoch % 1 == 0:
    print(f'epoch {epoch + 1} : w = {W : .3f}, loss = {loss : .8f}')
```
